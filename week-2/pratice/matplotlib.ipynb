```python
import pandas as pd
```

```python
housing = pd.read_csv("Housing.csv")
```

```python
housing.info()
```

```python
housing.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
housing.area.plot(kind='line', color='red',label='area',linewidth=5, alpha=1,grid=True, linestyle=':')
housing.price.plot(color='red',label='price', linewidth=2, alpha=1, grid=True,linestyle='--')
plt.legend(loc='upper right')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Housing Information')
plt.show()
```

```python
housing.plot(kind='scatter', x='area',y='price',alpha=0.4, color='r',label='stories')
plt.show()
```

```python
# Histogram

housing.price.plot(kind='hist', bins=50)
plt.show()
```
